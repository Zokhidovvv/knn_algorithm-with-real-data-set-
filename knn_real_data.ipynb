```python
import pandas as pd
import numpy as np

df = pd.read_csv('iris.csv')
df.head()
```

```python
X = df.drop(['Species','Id'], axis =1).values
y = df['Species']
x_new = np.array([5, 3, 1.4,0.2])

sample = df.shape[0]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
# 1. Fit va transform X_train
X_scaled = scaler.fit_transform(X)

# 2. Transform x_new
x_new_scaled = scaler.transform([x_new]) 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size = 0.2,
    random_state = 42,
    shuffle = True
        
)
print(X_train.shape, X_test.shape)   
```

```python
def euclidean(x1,x2):
    return np.sqrt(np.sum((x1 - x2)**2))
```

```python
from collections import Counter

from collections import Counter

def knn_predict(X_train, y_train, x_new, k):
    distances = []
    
    for i in range(len(X_train)):
        d = euclidean(X_train[i], x_new)  
        distances.append((d, y_train.iloc[i]))

    distances.sort(key=lambda x: x[0])

    k_nearest = distances[:k]

    votes = [label for dist, label in k_nearest]
    counts = Counter(votes)

    return counts.most_common(1)[0][0]

natija = knn_predict(X_train, y_train, x_new, k = 3)

if natija == 'Iris-setosa':
    print('Iris-setosa')
elif natija == 'Iris-virginica':
    print('Iris-virginica')
else:
    print('Iris-versicolor')
```
